```python
analyser=create_analyzer(task="sentiment", lang="es")

opiniones = [
    "La inteligencia artificial es el futuro y me encanta",
    "No entiendo nada de IA y me genera mucha frustración",
    "Las herramientas de IA son muy útiles para estudiar ingeniería",
    "Me da miedo que la IA reemplace mi trabajo pronto",
    "Python es el mejor lenguaje para aprender ciencia de datos",
    "Es muy difícil mantenerse al día con tanta tecnología nueva",
    "Amo los retos que propone la computación en la nube"
]
df=pd.DataFrame(opiniones,columns=['comentario'])
```

```python
def analizar_es(texto):
  resultado=analyser.predict(texto)
  mapping={'POS':'Positivo', 'NEU':'Neutral' , 'NEG':'Negativo'}
  return mapping[resultado.output]

df['resultado']=df['comentario'].apply(analizar_es)
df
```
